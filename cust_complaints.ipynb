```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
from datetime import datetime
from itertools import groupby
```

```python
def load_file():
    column_names = ['Complaint ID','Product','Sub-product','Issue','Sub-issue','State','ZIP code','Submitted via','Date received','Date sent to company','Company','Company response','Timely response?','Consumer disputed?']
    return pd.read_csv('complaints_dec_2014.csv')

```

```python
entire_cons_compl = load_file()

entire_cons_compl.head(6)
```

```python

prodID_index = entire_cons_compl.pop('Product')
entire_cons_compl.set_index(prodID_index)

num_complaints_by_prod = prodID_index.value_counts()
print(num_complaints_by_prod)

```

```python
comp_index = entire_cons_compl.pop('Company')
num_complaints_by_company = comp_index.value_counts()

print(num_complaints_by_company.head(10))

```

```python
complaint_dates = entire_cons_compl.pop('Date received')

entire_cons_compl.index = pd.to_datetime(complaint_dates)#, format='%m/%d/%Y')

entire_cons_compl['Weekday'] = entire_cons_compl.index.weekday

entire_cons_compl.head()


#DOC_list = day_of_complaint.weekday
```

```python
def date_to_check(each):
    day = datetime.strptime(each, '%m/%d/%Y')
    return day.weekday()
```

```python
dates_to_check = complaint_dates.value_counts().index
day_tally = np.array([0,0,0,0,0,0,0])

for each in dates_to_check:
    day = date_to_check(each)
    day_tally[day] += 1
    
print(day_tally)
```

```python
days=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

totals_for_Mon_thru_Sun = entire_cons_compl['Weekday'].value_counts()

day_of_week_of_totals = totals_for_Mon_thru_Sun.index

for i in range(len(totals_for_Mon_thru_Sun)):
    the_mean = totals_for_Mon_thru_Sun[i] / day_tally[i]
    print("Mean for {} is ".format(days[day_of_week_of_totals[i]]),end='')
    print(the_mean)
```
